```python
def calc_running_totals(df):
    rolling_avg_possession = []
    rolling_avg_Ortg = []
    rolling_avg_Drtg = []
    rolling_avg_Net_rtg = []
    rolling_avg_EFG = []
    rolling_avg_TS = []
    rolling_avg_3Ppct = []
    rolling_avg_3PAP48 = []
    rolling_avg_FTpct = []
    rolling_avg_FTAP48 = []
    rolling_avg_FTR = []
    rolling_avg_ORBR = []
    rolling_avg_DRBR = []
    rolling_avg_AP48 = []
    rolling_avg_TOP48 = []
    rolling_avg_ATORatio = []
    rolling_win_pct = []
    rolling_opp_EFG = []
    rolling_opp_TS = []
    rolling_SP48 = []
    rolling_BP48 = []
    
    L10_avg_possession = []
    L10_avg_Ortg = []
    L10_avg_Drtg = []
    L10_avg_Net_rtg = []
    L10_avg_EFG = []
    L10_avg_TS = []
    L10_avg_3Ppct = []
    L10_avg_3PAP48 = []
    L10_avg_FTpct = []
    L10_avg_FTAP48 = []
    L10_avg_FTR = []
    L10_avg_ORBR = []
    L10_avg_DRBR = []
    L10_avg_AP48 = []
    L10_avg_TOP48 = []
    L10_avg_ATORatio = []
    L10_win_pct = []
    L10_opp_EFG = []
    L10_opp_TS = []
    L10_SP48 = []
    L10_BP48 = []

    for index, row in df.iterrows():
        team_games = df[df['teamAbbr']== row['teamAbbr']]
        team_games = team_games.reset_index(drop = True)
        #print(team_games)
        if row['gmDate'] == team_games['gmDate'][0]:
            rolling_avg_possession.append(row['poss'])
            rolling_avg_Ortg.append(row['teamOrtg'])
            rolling_avg_Drtg.append(row['teamDrtg'])
            rolling_avg_Net_rtg.append(row['teamOrtg']-row['teamDrtg'])
            rolling_avg_EFG.append((row['teamFGM']+0.5*row['team3PM'])/row['teamFGA']) 
            TSA = row['teamFGA'] + (0.44*row['teamFTA'])
            rolling_avg_TS.append(row['teamPTS']/(2*TSA))
            rolling_avg_3Ppct.append(row['team3PM']/row['team3PA'])
            rolling_avg_3PAP48.append((row['team3PA']/row['teamMin'])*48*5)
            rolling_avg_FTpct.append(row['teamFTM']/row['teamFTA'])
            rolling_avg_FTAP48.append((row['teamFTA']/row['teamMin'])*48*5)
            rolling_avg_FTR.append(row['teamFTA']/row['teamFGA'])
            rolling_avg_ORBR.append(row['teamORB']/(row['teamORB']+row['opptDRB']))
            rolling_avg_DRBR.append(row['teamDRB']/(row['teamDRB']+row['opptORB']))
            rolling_avg_AP48.append((row['teamAST']/row['teamMin'])*48*5)
            rolling_avg_TOP48.append((row['teamTO']/row['teamMin'])*48*5)
            rolling_avg_ATORatio.append(row['teamAST']/row['teamTO'])
            rolling_win_pct.append(0)
            rolling_opp_EFG.append((row['opptFGM']+0.5*row['oppt3PM'])/row['opptFGA'])
            opptTSA = row['opptFGA']+(0.44*row['opptFTA'])
            rolling_opp_TS.append(row['opptPTS']/(2*opptTSA))
            rolling_SP48.append((row['teamSTL']/row['teamMin'])*48*5)
            rolling_BP48.append((row['teamBLK']/row['teamMin'])*48*5)
            
            L10_avg_possession.append(row['poss'])
            L10_avg_Ortg.append(row['teamOrtg'])
            L10_avg_Drtg.append(row['teamDrtg'])
            L10_avg_Net_rtg.append(row['teamOrtg']-row['teamDrtg'])
            L10_avg_EFG.append((row['teamFGM']+0.5*row['team3PM'])/row['teamFGA'])
            L10_avg_TS.append(row['teamPTS']/(2*TSA))
            L10_avg_3Ppct.append(row['team3PM']/row['team3PA'])
            L10_avg_3PAP48.append((row['team3PA']/row['teamMin'])*48*5)
            L10_avg_FTpct.append(row['teamFTM']/row['teamFTA'])
            L10_avg_FTAP48.append((row['teamFTA']/row['teamMin'])*48*5)
            L10_avg_FTR.append(row['teamFTA']/row['teamFGA'])
            L10_avg_ORBR.append(row['teamORB']/(row['teamORB']+row['opptDRB']))
            L10_avg_DRBR.append(row['teamDRB']/(row['teamDRB']+row['opptORB']))
            L10_avg_AP48.append((row['teamAST']/row['teamMin'])*48*5)
            L10_avg_TOP48.append((row['teamTO']/row['teamMin'])*48*5)
            L10_avg_ATORatio.append(row['teamAST']/row['teamTO'])
            L10_win_pct.append(0)
            L10_opp_EFG.append((row['opptFGM']+0.5*row['oppt3PM'])/row['opptFGA'])
            L10_opp_TS.append(row['opptPTS']/(2*opptTSA))
            L10_SP48.append((row['teamSTL']/row['teamMin'])*48*5)
            L10_BP48.append((row['teamBLK']/row['teamMin'])*48*5)
        else:
            previous_games = team_games[team_games['gmDate'] < row['gmDate']]
            #print(previous_games)
            rolling_avg_possession.append(stat.mean(previous_games['poss']))
            rolling_avg_Ortg.append(stat.mean(previous_games['teamOrtg']))
            rolling_avg_Drtg.append(stat.mean(previous_games['teamDrtg']))
            rolling_avg_Net_rtg.append(stat.mean(previous_games['teamOrtg'])-stat.mean(previous_games['teamDrtg']))
            total_FGM = previous_games['teamFGM'].sum()
            total_FTM = previous_games['teamFTM'].sum()
            total_3PM = previous_games['team3PM'].sum()
            total_FGA = previous_games['teamFGA'].sum()
            total_FTA = previous_games['teamFTA'].sum()
            total_3PA = previous_games['team3PA'].sum()
            total_TSA = total_FGA + (0.44*total_FTA)
            total_OREB = previous_games['teamORB'].sum()
            total_DREB = previous_games['teamDRB'].sum()
            total_Opp_OREB = previous_games['opptORB'].sum()
            total_Opp_DREB = previous_games['opptDRB'].sum()
            total_points = previous_games['teamPTS'].sum()
            total_AST = previous_games['teamAST'].sum()
            total_TO = previous_games['teamTO'].sum()
            total_STL = previous_games['teamSTL'].sum()
            total_BLK = previous_games['teamBLK'].sum()
            total_Min = previous_games['teamMin'].sum()
            total_wins = (previous_games['teamRslt']=='Win').sum()
            total_games = len(previous_games)
            opp_total_FGM = previous_games['opptFGM'].sum()
            opp_total_3PM = previous_games['oppt3PM'].sum()
            opp_total_3PA = previous_games['oppt3PA'].sum()
            opp_total_FGA = previous_games['opptFGA'].sum()
            opp_total_FTA = previous_games['opptFTA'].sum()
            opp_total_TSA = opp_total_FGA + (0.44*opp_total_FTA)
            opp_total_points = previous_games['opptPTS'].sum()
            rolling_avg_EFG.append((total_FGM+0.5*total_3PM)/total_FGA)
            rolling_avg_TS.append(total_points/(2*total_TSA))
            rolling_avg_3Ppct.append(total_3PM/total_3PA)
            rolling_avg_3PAP48.append((total_3PA/total_Min)*48*5)
            rolling_avg_FTpct.append(total_FTM/total_FTA)
            rolling_avg_FTAP48.append((total_FTA/total_Min)*48*5)
            rolling_avg_FTR.append(total_FTA/total_FGA)
            rolling_avg_ORBR.append(total_OREB/(total_OREB+total_Opp_DREB))
            rolling_avg_DRBR.append(total_DREB/(total_DREB+total_Opp_OREB))
            rolling_avg_AP48.append((total_AST/total_Min)*48*5)
            rolling_avg_TOP48.append((total_TO/total_Min)*48*5)
            rolling_avg_ATORatio.append(total_AST/total_TO)
            rolling_win_pct.append(total_wins/total_games)
            rolling_opp_EFG.append((opp_total_FGM + 0.5*opp_total_3PM)/opp_total_FGA)
            rolling_opp_TS.append(opp_total_points/(2*opp_total_TSA))
            rolling_SP48.append((total_STL/total_Min)*48*5)
            rolling_BP48.append((total_BLK/total_Min)*48*5)
            
            if len(previous_games) < 10:
                L10_games = previous_games.tail(len(previous_games))
         
                L10_avg_possession.append(stat.mean(L10_games['poss']))
                L10_avg_Ortg.append(stat.mean(L10_games['teamOrtg']))
                L10_avg_Drtg.append(stat.mean(L10_games['teamDrtg']))
                L10_avg_Net_rtg.append(stat.mean(L10_games['teamOrtg']) - stat.mean(L10_games['teamDrtg']))
                total_FGM = L10_games['teamFGM'].sum()
                total_FTM = L10_games['teamFTM'].sum()
                total_3PM = L10_games['team3PM'].sum()
                total_FGA = L10_games['teamFGA'].sum()
                total_FTA = L10_games['teamFTA'].sum()
                total_3PA = L10_games['team3PA'].sum()
                total_TSA = total_FGA + (0.44*total_FTA)
                total_OREB = L10_games['teamORB'].sum()
                total_DREB = L10_games['teamDRB'].sum()
                total_Opp_OREB = L10_games['opptORB'].sum()
                total_Opp_DREB = L10_games['opptDRB'].sum()
                total_points = L10_games['teamPTS'].sum()
                total_AST = L10_games['teamAST'].sum()
                total_TO = L10_games['teamTO'].sum()
                total_STL = L10_games['teamSTL'].sum()
                total_BLK = L10_games['teamBLK'].sum()
                total_Min = L10_games['teamMin'].sum()
                total_wins = (L10_games['teamRslt']=='Win').sum()
                total_games = 10
                opp_total_FGM = L10_games['opptFGM'].sum()
                opp_total_3PM = L10_games['oppt3PM'].sum()
                opp_total_3PA = L10_games['oppt3PA'].sum()
                opp_total_FGA = L10_games['opptFGA'].sum()
                opp_total_FTA = L10_games['opptFTA'].sum()
                opp_total_TSA = opp_total_FGA + (0.44*opp_total_FTA)
                opp_total_points = L10_games['opptPTS'].sum()
                L10_avg_EFG.append((total_FGM+0.5*total_3PM)/total_FGA)
                L10_avg_TS.append(total_points/(2*total_TSA))
                L10_avg_3Ppct.append(total_3PM/total_3PA)
                L10_avg_3PAP48.append((total_3PA/total_Min)*48*5)
                L10_avg_FTpct.append(total_FTM/total_FTA)
                L10_avg_FTAP48.append((total_FTA/total_Min)*48*5)
                L10_avg_FTR.append(total_FTA/total_FGA)
                L10_avg_ORBR.append(total_OREB/(total_OREB+total_Opp_DREB))
                L10_avg_DRBR.append(total_DREB/(total_DREB+total_Opp_OREB))
                L10_avg_AP48.append((total_AST/total_Min)*48*5)
                L10_avg_TOP48.append((total_TO/total_Min)*48*5)
                L10_avg_ATORatio.append(total_AST/total_TO)
                L10_win_pct.append(total_wins/total_games)
                L10_opp_EFG.append((opp_total_FGM + 0.5*opp_total_3PM)/opp_total_FGA)
                L10_opp_TS.append(opp_total_points/(2*opp_total_TSA))
                L10_SP48.append((total_STL/total_Min)*48*5)
                L10_BP48.append((total_BLK/total_Min)*48*5)
            else:
                L10_games = previous_games.tail(10)
                L10_avg_possession.append(stat.mean(L10_games['poss']))
                L10_avg_Ortg.append(stat.mean(L10_games['teamOrtg']))
                L10_avg_Drtg.append(stat.mean(L10_games['teamDrtg']))
                L10_avg_Net_rtg.append(stat.mean(L10_games['teamOrtg']) - stat.mean(L10_games['teamDrtg']))
                total_FGM = L10_games['teamFGM'].sum()
                total_FTM = L10_games['teamFTM'].sum()
                total_3PM = L10_games['team3PM'].sum()
                total_FGA = L10_games['teamFGA'].sum()
                total_FTA = L10_games['teamFTA'].sum()
                total_3PA = L10_games['team3PA'].sum()
                total_TSA = total_FGA + (0.44*total_FTA)
                total_OREB = L10_games['teamORB'].sum()
                total_DREB = L10_games['teamDRB'].sum()
                total_Opp_OREB = L10_games['opptORB'].sum()
                total_Opp_DREB = L10_games['opptDRB'].sum()
                total_points = L10_games['teamPTS'].sum()
                total_AST = L10_games['teamAST'].sum()
                total_TO = L10_games['teamTO'].sum()
                total_STL = L10_games['teamSTL'].sum()
                total_BLK = L10_games['teamBLK'].sum()
                total_Min = L10_games['teamMin'].sum()
                total_wins = (L10_games['teamRslt']=='Win').sum()
                total_games = 10
                opp_total_FGM = L10_games['opptFGM'].sum()
                opp_total_3PM = L10_games['oppt3PM'].sum()
                opp_total_3PA = L10_games['oppt3PA'].sum()
                opp_total_FGA = L10_games['opptFGA'].sum()
                opp_total_FTA = L10_games['opptFTA'].sum()
                opp_total_TSA = opp_total_FGA + (0.44*opp_total_FTA)
                opp_total_points = L10_games['opptPTS'].sum()
                L10_avg_EFG.append((total_FGM+0.5*total_3PM)/total_FGA)
                L10_avg_TS.append(total_points/(2*total_TSA))
                L10_avg_3Ppct.append(total_3PM/total_3PA)
                L10_avg_3PAP48.append((total_3PA/total_Min)*48*5)
                L10_avg_FTpct.append(total_FTM/total_FTA)
                L10_avg_FTAP48.append((total_FTA/total_Min)*48*5)
                L10_avg_FTR.append(total_FTA/total_FGA)
                L10_avg_ORBR.append(total_OREB/(total_OREB+total_Opp_DREB))
                L10_avg_DRBR.append(total_DREB/(total_DREB+total_Opp_OREB))
                L10_avg_AP48.append((total_AST/total_Min)*48*5)
                L10_avg_TOP48.append((total_TO/total_Min)*48*5)
                L10_avg_ATORatio.append(total_AST/total_TO)
                L10_win_pct.append(total_wins/total_games)
                L10_opp_EFG.append((opp_total_FGM + 0.5*opp_total_3PM)/opp_total_FGA)
                L10_opp_TS.append(opp_total_points/(2*opp_total_TSA))
                L10_SP48.append((total_STL/total_Min)*48*5)
                L10_BP48.append((total_BLK/total_Min)*48*5)

            
    df['avg_poss']=rolling_avg_possession
    df['Off_Eff']=rolling_avg_Ortg
    df['Def_Eff']=rolling_avg_Drtg
    df['Net_Eff']=rolling_avg_Net_rtg
    df['Prev_EFG'] = rolling_avg_EFG
    df['Prev_TS'] = rolling_avg_TS
    df['Prev_3Ppct'] = rolling_avg_3Ppct
    df['Prev_3PAP48'] = rolling_avg_3PAP48
    df['Prev_FTpct'] = rolling_avg_FTpct
    df['Prev_FTAP48'] = rolling_avg_FTAP48
    df['Prev_FTR'] = rolling_avg_FTR
    df['ORBR'] = rolling_avg_ORBR
    df['DRBR'] = rolling_avg_DRBR
    df['AP48'] = rolling_avg_AP48
    df['TOP48'] = rolling_avg_TOP48
    df['ATORatio'] = rolling_avg_ATORatio
    df['Win_Pct'] = rolling_win_pct
    df['Opp_EFG'] = rolling_opp_EFG
    df['Opp_TS'] = rolling_opp_TS
    df['SP48'] = rolling_SP48
    df['BP48'] = rolling_BP48
    
    df['L10_avg_poss'] = L10_avg_possession
    df['L10_Off_Eff'] = L10_avg_Ortg
    df['L10_Def_Eff'] = L10_avg_Drtg
    df['L10_Net_Eff'] = L10_avg_Net_rtg
    df['L10_Prev_EFG'] = L10_avg_EFG
    df['L10_Prev_TS'] = L10_avg_TS
    df['L10_3Ppct'] = L10_avg_3Ppct
    df['L10_3PAP48'] = L10_avg_3PAP48
    df['L10_FTpct'] = L10_avg_FTpct
    df['L10_FTAP48'] = L10_avg_FTAP48
    df['L10_FTR'] = L10_avg_FTR
    df['L10_ORBR'] = L10_avg_ORBR
    df['L10_DRBR'] = L10_avg_DRBR
    df['L10_AP48'] = L10_avg_AP48
    df['L10_TOP48'] = L10_avg_TOP48
    df['L10_ATORatio'] = L10_avg_ATORatio
    df['L10_Win_Pct'] = L10_win_pct
    df['L10_Opp_EFG'] = L10_opp_EFG
    df['L10_Opp_TS'] = L10_opp_TS
    df['L10_SP48'] = L10_SP48
    df['L10_BP48'] = L10_BP48
```

```python
for i in box_scores_seasons:
    calc_running_totals(i)
    i['MOV'] =  np.where(i['teamLoc'] == 'Home',
                                   i['teamPTS'] - i['opptPTS'],
                                   i['opptPTS'] - i['teamPTS'])  #Calculate Margin of Victory for each home team
    i['Official1'] = i['offFNm1'].astype(str) + ' ' + i['offLNm1'].astype(str) #Combine official's first and last name
    i['Official2'] = i['offFNm2'].astype(str) + ' ' + i['offLNm2'].astype(str)
    i['Official3'] = i['offFNm3'].astype(str) + ' ' + i['offLNm3'].astype(str)
    
```
